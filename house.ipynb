```python
### split data ###
X=df.drop('price',axis=1)
y=df['price']
```

```python
print("X of shape",X.shape)
print("y of shape",y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=50)
```

```python
print('Shape of X_train = ', X_train.shape)
print('Shape of y_train = ', y_train.shape)
print('Shape of X_test = ', X_test.shape)
print('Shape of y_test = ', y_test.shape)
 
```

```python
    ### Feature Scalling ###
    from sklearn.preprocessing import StandardScaler
    sc=StandardScaler()
    sc.fit(X_train)
    X_train=sc.transform(X_train)
    X_test=sc.transform(X_test)
```

```python
## Linear regression ##
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
### coffecient of features models
lr.coef_
```

```python
### baseline of dependent and indepedent variable
lr.intercept_
```

```python
## predict the value of home and test it
X_test[0, :]#first row print
```

```python
lr.predict([X_test[0, :]])
```

```python
lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```

```python
from sklearn .linear_model import Ridge, Lasso
```

```python
rd=Ridge()
rd.fit(X_train,y_train)
rd.score(X_test,y_test)
```

```python
ls=Lasso()
ls.fit(X_train,y_train)
ls.score(X_test,y_test)
```

```python
ls3=Lasso(alpha=3)
ls3.fit(X_train,y_train)
ls3.score(X_test,y_test)
```
